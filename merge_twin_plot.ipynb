```python
import pandas as pd
import matplotlib.pyplot as plt
import os

folder_path = "/Users/tic/Desktop/tracking_data/"
file_path = "merged_2025_07_28_11M8-3_clip30.csv"
data = os.path.join(folder_path, file_path)
 
df_merged = pd.read_csv(data)

df_merged['Mid_Ear_x'] = (df_merged['Left_Ear_x'] + df_merged['Right_Ear_x']) / 2
df_merged.head()
```

```python
fig, ax1 = plt.subplots(figsize=(20, 6))

color1 = "green"
ax1.set_xlabel("Time(seconds)", fontsize=20)
ax1.set_ylabel("dF/F(%)", fontsize=20, color=color1)
ax1.plot(df_merged["Time_seconds"], df_merged["dF_F (%)"], color=color1, label="dF/F(%)")
ax1.tick_params(axis='y', labelsize=20, labelcolor=color1)
ax1.tick_params(axis='x', labelsize=18)


ax2 = ax1.twinx()
color2 = "blue"
ax2.set_ylabel("X_postion(Mid Ear)", fontsize=20 , color=color2)
ax2.plot(df_merged["Time_seconds"], df_merged["Mid_Ear_x"],color=color2, label="X position")
ax2.tick_params(axis='y', labelsize=18, labelcolor=color2)
ax2.set_yticks([200, 400, 600, 800, 1000])

fig.tight_layout()
plt.title("Neural Activity(dF/F) and Position All", fontsize=20)

plt.show()
```
